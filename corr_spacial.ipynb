```python
difference = velib_nombre.diff(axis=1)
```

```python
series = difference[difference.columns[2:]].abs().sum(axis = 1)
```

```python
series = (series - series.mean())/series.std()
```

```python
station_w_number = pd.concat([velib_nombre.station_id, series], axis=1)
```

```python
import urllib.request
import json
from pandas.io.json import json_normalize
```

```python
request= urllib.request.Request('https://velib-metropole-opendata.smoove.pro/opendata/Velib_Metropole/station_information.json')
response = urllib.request.urlopen(request)
station_info = response.read()
data = json.loads(station_info)
df_station_info = json_normalize(data['data']['stations'])
df_station_info.head()
```

```python
velib = pd.merge(df_station_info[["station_id", "lat", "lon"]], station_w_number, left_on = "station_id", right_on = "station_id")
```

```python
graph = {}
for i, j in enumerate(velib.station_id) :
    graph[j] = [[ velib.iloc[i][0], velib.iloc[i]['lat'], velib.iloc[i]['lon']],[]]

def distance (graph, i, j):
  return (abs(graph[i][0][1] - graph[j][0][1]) ** 2 + abs(graph[i][0][2] - graph[i][0][2])** 2) ** (1 / 2)

for i in graph.keys():
    for j in graph.keys(): 
        graph[i][1].append([j,distance(graph,i ,j)])
```

```python
listes = {}
numbers = [j for j in range(2, 200)]
for j in numbers:
  listes[j] = []
  for i in graph.keys():
    graph[i][1].sort(key = lambda x : x[1])
    graph[i][0].append(statistics.mean([graph[values[0]][0][0] for values in graph[i][1][1:j]]))
    listes[j].append([graph[i][0][0], graph[i][0][3 + j - 2]])
```

```python
from scipy import stats
```

```python
X = {}
Y = {}
list_corr = []
for j in numbers:
  X[j] = [values[0] for values in listes[j]]
  Y[j] = [values[1] for values in listes[j]]
  list_corr.append(stats.pearsonr(X[j], Y[j])[0])
```

```python
sns.lineplot(numbers, list_corr)
```

```python
import requests
import tempfile
import zipfile

url = 'https://www.data.gouv.fr/fr/datasets/r/07b7c9a2-d1e2-4da6-9f20-01a7b72d4b12'
temporary_location = tempfile.gettempdir()

def download_unzip(url, dirname = tempfile.gettempdir(), destname = "borders"):
    myfile = requests.get(url)
    open(dirname + '/' + destname + '.zip', 'wb').write(myfile.content)
    with zipfile.ZipFile(dirname + '/' + destname + '.zip', 'r') as zip_ref:
        zip_ref.extractall(dirname + '/' + destname)
        
download_unzip(url)
communes = gpd.read_file(temporary_location + "/borders/communes-20190101.json")

communes_borders = gpd.read_file(temporary_location + "/borders/communes-20190101.json")
communes_borders.crs

communes['dep'] = communes.insee.str[:2]
```
